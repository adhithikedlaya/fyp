# Parameter sweep to determine which hemodynamic P-DCM parameters affect CBF and CBV the most

```python
from Euler_1stOrderForward import getEulerFandV
import numpy as np
import matplotlib.pyplot as plt


rng = np.arange(0, 2.0, 0.5)
psi = 0.6 #affects f and v
phi = 1.5
chi = 0.6
tMTT = 2.0
tau = 4.0
alpha = 0.32
E0 = 0.4
amps_f = []
amps_v = []
num_sims = 50

std_devs_f = []
std_devs_v = []

for inc in rng:
    all_fs = []
    all_vs = []
    for i in range(num_sims):
        t, fs, vs = getEulerFandV(length=1000, psi=psi+inc, phi=phi, chi=chi, tMTT=tMTT, tau=tau, alpha=alpha, E0=E0, noise=True)
        all_fs.append(fs.detach().numpy())
        all_vs.append(vs.detach().numpy())
    std_dev_f = np.max(np.nanstd(all_fs, axis=0))
    std_dev_v = np.max(np.nanstd(all_vs, axis=0))
    print(std_dev_v)
    std_devs_f.append(std_dev_f)
    std_devs_v.append(std_dev_v)

plt.plot(rng, np.array(std_devs_f), label='CBF')
plt.plot(rng, np.array(std_devs_v), label='CBV')
plt.legend()
plt.xlabel("Psi")
plt.ylabel("Max standard deviation")
plt.title(f"How CBF and CBV vary with psi noise")
plt.show()

std_devs_f = []
std_devs_v = []

for inc in rng:
    all_fs = []
    all_vs = []
    for i in range(num_sims):
        t, fs, vs = getEulerFandV(length=1000, psi=psi, phi=phi+inc, chi=chi, tMTT=tMTT, tau=tau, alpha=alpha, E0=E0, noise=True)
        all_fs.append(fs.detach().numpy())
        all_vs.append(vs.detach().numpy())
    std_dev_f = np.max(np.nanstd(all_fs, axis=0))
    std_dev_v = np.max(np.nanstd(all_vs, axis=0))
    print(std_dev_v)
    std_devs_f.append(std_dev_f)
    std_devs_v.append(std_dev_v)

plt.plot(rng, np.array(std_devs_f), label='CBF')
plt.plot(rng, np.array(std_devs_v), label='CBV')
plt.legend()
plt.xlabel("Phi")
plt.ylabel("Max standard deviation")
plt.title(f"How CBF and CBV vary with phi noise")
plt.show()

std_devs_f = []
std_devs_v = []

for inc in rng:
    all_fs = []
    all_vs = []
    for i in range(num_sims):
        t, fs, vs = getEulerFandV(length=1000, psi=psi, phi=phi, chi=chi+inc, tMTT=tMTT, tau=tau, alpha=alpha, E0=E0, noise=True)
        all_fs.append(fs.detach().numpy())
        all_vs.append(vs.detach().numpy())
    std_dev_f = np.max(np.nanstd(all_fs, axis=0))
    std_dev_v = np.max(np.nanstd(all_vs, axis=0))
    print(std_dev_v)
    std_devs_f.append(std_dev_f)
    std_devs_v.append(std_dev_v)

plt.plot(rng, np.array(std_devs_f), label='CBF')
plt.plot(rng, np.array(std_devs_v), label='CBV')
plt.legend()
plt.xlabel("chi")
plt.ylabel("Max standard deviation")
plt.title(f"How CBF and CBV vary with chi (resting state)")
plt.show()

std_devs_f = []
std_devs_v = []

for inc in rng:
    all_fs = []
    all_vs = []
    for i in range(num_sims):
        t, fs, vs = getEulerFandV(length=1000, psi=psi, phi=phi, chi=chi, tMTT=tMTT+inc, tau=tau, alpha=alpha, E0=E0, noise=True)
        all_fs.append(fs.detach().numpy())
        all_vs.append(vs.detach().numpy())
    std_dev_f = np.max(np.nanstd(all_fs, axis=0))
    std_dev_v = np.max(np.nanstd(all_vs, axis=0))
    print(std_dev_v)
    std_devs_f.append(std_dev_f)
    std_devs_v.append(std_dev_v)

plt.plot(rng, np.array(std_devs_f), label='CBF')
plt.plot(rng, np.array(std_devs_v), label='CBV')
plt.legend()
plt.xlabel("mean transit time")
plt.ylabel("Max standard deviation")
plt.title(f"How CBF and CBV vary with mean transit time (resting state)")
plt.show()

std_devs_f = []
std_devs_v = []

for inc in rng:
    all_fs = []
    all_vs = []
    for i in range(num_sims):
        t, fs, vs = getEulerFandV(length=1000, psi=psi, phi=phi, chi=chi, tMTT=tMTT, tau=tau+inc, alpha=alpha, E0=E0, noise=True)
        all_fs.append(fs.detach().numpy())
        all_vs.append(vs.detach().numpy())
    std_dev_f = np.max(np.nanstd(all_fs, axis=0))
    std_dev_v = np.max(np.nanstd(all_vs, axis=0))
    print(std_dev_v)
    std_devs_f.append(std_dev_f)
    std_devs_v.append(std_dev_v)

plt.plot(rng, np.array(std_devs_f), label='CBF')
plt.plot(rng, np.array(std_devs_v), label='CBV')
plt.legend()
plt.xlabel("tau")
plt.ylabel("Max standard deviation")
plt.title(f"How CBF and CBV vary with tau (resting state)")
plt.show()

std_devs_f = []
std_devs_v = []

for inc in rng:
    all_fs = []
    all_vs = []
    for i in range(num_sims):
        t, fs, vs = getEulerFandV(length=1000, psi=psi, phi=phi, chi=chi, tMTT=tMTT, tau=tau, alpha=alpha+inc, E0=E0, noise=True)
        all_fs.append(fs.detach().numpy())
        all_vs.append(vs.detach().numpy())
    std_dev_f = np.max(np.nanstd(all_fs, axis=0))
    std_dev_v = np.max(np.nanstd(all_vs, axis=0))
    print(std_dev_v)
    std_devs_f.append(std_dev_f)
    std_devs_v.append(std_dev_v)

plt.plot(rng, np.array(std_devs_f), label='CBF')
plt.plot(rng, np.array(std_devs_v), label='CBV')
plt.legend()
plt.xlabel("alpha")
plt.ylabel("Max standard deviation")
plt.title(f"How CBF and CBV vary with alpha (resting state)")
plt.show()

std_devs_f = []
std_devs_v = []

for inc in rng:
    all_fs = []
    all_vs = []
    for i in range(num_sims):
        t, fs, vs = getEulerFandV(length=1000, psi=psi, phi=phi, chi=chi, tMTT=tMTT, tau=tau, alpha=alpha, E0=E0+inc, noise=True)
        all_fs.append(fs.detach().numpy())
        all_vs.append(vs.detach().numpy())
    std_dev_f = np.max(np.nanstd(all_fs, axis=0))
    std_dev_v = np.max(np.nanstd(all_vs, axis=0))
    print(std_dev_v)
    std_devs_f.append(std_dev_f)
    std_devs_v.append(std_dev_v)

plt.plot(rng, np.array(std_devs_f), label='CBF')
plt.plot(rng, np.array(std_devs_v), label='CBV')
plt.legend()
plt.xlabel("E0")
plt.ylabel("Max standard deviation")
plt.title(f"How CBF and CBV vary with E0(resting state)")
plt.show()
```
